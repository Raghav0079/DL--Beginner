```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("uciml/pima-indians-diabetes-database")

print("Path to dataset files:", path)
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(path + '/diabetes.csv')
print(df.head())
```

```python
df.corr()['Outcome']
```

```python
x = df.iloc[: , : -1].values
y = df.iloc[: , -1].values
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(x)
```

```python
X
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(32, activation = 'relu',input_dim = 8))
model.add(Dense(1, activation = 'relu'))

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs = 10, batch_size = 32,validation_data=(X_test, y_test))
```

```python
import keras_tuner as kt
```

```python
def build_model(hp):
    model = Sequential()
    model.add(Dense(32,activation='relu',input_dim=8))
    model.add(Dense(1,activation='sigmoid'))
    
    model.compile(optimizer = hp.Choice('optimizer',['adam','rmsprop','sgd','adadelta']), loss = 'binary_crossentropy', metrics = ['accuracy'])
    
    return model
```

```python
tuner = kt.RandomSearch(build_model,
                        objective = 'val_accuracy',
                        max_trials=5)
```

```python
tuner.search(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model = tuner.get_best_models(num_models=1)[0]
```

```python
model.summary()
```

```python
model.fit(X_train, y_train,batch_size=32, epochs=100,initial_epoch=6, validation_data=(X_test, y_test))
```

```python
def build_model(hp):
    model = Sequential()
    units = hp.Int("units", 8, 128, step=8)

    model.add(Dense(units=units , activation = 'relu',input_dim = 8 ))
    model.add(Dense(1, activation = 'sigmoid'))

    model.compile(optimizer="rmsprop", loss="binary_crossentropy", metrics=["accuracy"])
    return model
```

```python
tuner = kt.RandomSearch(build_model,
                        objective = 'val_accuracy',
                        max_trials=5)
```

```python
tuner.search(X_train, y_train, epochs=5, validation_data=(X_test, y_test)) 
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
# Clear the tuner directory to avoid model loading conflicts
import shutil
import os

# Create a new tuner with a unique project name to avoid conflicts
tuner_new = kt.RandomSearch(build_model,
						   objective='val_accuracy',
						   max_trials=5,
						   project_name='diabetes_tuning_v2')

# Search for best hyperparameters
tuner_new.search(X_train, y_train, epochs=5, validation_data=(X_test, y_test))

# Get the best model from the new tuner
model = tuner_new.get_best_models(num_models=1)[0]
```

```python
model.fit(X_train, y_train, epochs = 100, batch_size = 10,initial_epoch=6)
```

```python
def build_model(hp):
    
    model = Sequential()
    
    model.add(Dense(72,activation='relu',input_dim=8))
    for i in range(hp.Int('num_layers', min_value =1, max_value=10)):
        
        model.add(Dense(72,activation ='relu'))
    model.add(Dense(1,activation='sigmoid'))
    
    model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
    
    return model
```

```python
tuner = kt.RandomSearch(build_model,
                        objective='val_accuracy',
                        max_trials=3)
```

```python
tuner.search(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

```python
# Create a new tuner with a unique project name to avoid model loading conflicts
tuner_layers = kt.RandomSearch(build_model,
                               objective='val_accuracy',
                               max_trials=3,
                               project_name='diabetes_layers_tuning')

# Search for best hyperparameters
tuner_layers.search(X_train, y_train, epochs=5, validation_data=(X_test, y_test))

# Get the best model from the new tuner
model = tuner_layers.get_best_models(num_models=1)[0]
```

```python
model.fit(X_train, y_train, epochs = 100, batch_size = 10, initial_epoch=6, validation_data=(X_test, y_test))
```
